```python
# import Seq
from Bio.Seq import Seq

# import SeqIO
from Bio import SeqIO
```

```python
# clean sequence:
# "BDEFHIJKLMNOPQRSUVWXYZ"
def clean_Sequence(seq):
    for letter in "BDEFHIJKLMNOPQRSUVWXYZ":
        seq = seq.replace(letter,"")
    return seq
```

```python
# SeqIO.parse()
gb_iterator = SeqIO.parse('S5lab/sequence.fasta','fasta')
print(gb_iterator)

for record in gb_iterator:
    record.seq = Seq(clean_Sequence(str(record.seq)))
    print(record)
    print(record.id)
    print(record.description)
    print(record.seq)
```

### 2. Sequence Analysis and Plotting

```python
# Determine the molecular weight of both sequence
from Bio.SeqUtils import molecular_weight as mw

print("sA =",'%0.2f' % mw(SeqA))
print("sB =",'%0.2f' % mw(SeqB))
```

```python
# Refer to the Wallace rule to calculate the melting point of 
# sequence A and sequence B and plot them using Matplotlib library
from Bio.SeqUtils import MeltingTemp as mt

print('%0.2f' % mt.Tm_Wallace(SeqA))
print('%0.2f' % mt.Tm_Wallace(SeqB))

Tm_A = mt.Tm_Wallace(SeqA)
Tm_B = mt.Tm_Wallace(SeqB)
sequences = ['SeqA', 'SeqB']
melting_points = [Tm_A, Tm_B]

import matplotlib.pyplot as plt
plt.bar(sequences, melting_points, color=['blue', 'orange'])
plt.xlabel('Sequences')
plt.ylabel('Melting Temperature (Tm) °C')
plt.title('Melting Temperature of Sequences A and B')
plt.show()
```

```python
def calculateGC(seq):
    GC = []
    for x in seq:
        if x == "G" or x == "C":
            GC.append(x)
    return len(GC)/len(seq)

def calculateAT(seq):
    GC = []
    for x in seq:
        if x == "A" or x == "T":
            GC.append(x)
    return len(GC)/len(seq)
```

```python
# Calculate the GC and AT content of sequence A and sequence B.
print("GC content            :", calculateGC(SeqA))
percentageGC = calculateGC(SeqA) * 100
print("GC content percentage :", percentageGC)
```

```python
from Bio.SeqUtils import GC123
GCSeqA = GC123(SeqA)[0]
ATSeqA = 100- GCSeqA
GCSeqB = GC123(SeqB)[0]
ATSeqB = 100- GCSeqB
```

```python
# Plot the GC and AT content of sequence A and sequence B using Matplotlib library.
sequences = ['GCA', 'ATA','GCB','ATB']
ATGCPlot = [GCSeqA,ATSeqA,GCSeqB,ATSeqB]
plt.bar(sequences,ATGCPlot)
```

```python
# Plot the nucleotide base (A, C, G, T) frequency of
# sequence C using the Matplotlib library

from collections import Counter
freqA = Counter(SeqC)
print(freqA)

freqKey = freqA.keys()
freqVal = freqA.values()
print(freqKey)
print(freqVal)

import matplotlib.pyplot as plt
plt.bar(freqKey,freqVal)
plt.title("Nucleotide freq")
plt.show()
```

```python
# Transcribe the DNA sequence A and sequence B into mRNA.
mRNAseqA = SeqA.transcribe()
print(mRNAseqA)
mRNAseqB = SeqB.transcribe()
print(mRNAseqB)
```

```python
# Translate the mRNA sequence A and sequence B into amino acids sequences
aminoAcidSeqA = mRNAseqA.translate()
print(aminoAcidSeqA)
aminoAcidSeqB = mRNAseqB.translate()
print(aminoAcidSeqB)
```

```python
# Transcribe back the mRNA sequence A and sequence B into DNA.
DNAseqA = mRNAseqA.back_transcribe()
print(DNAseqA)
DNAseqB = mRNAseqB.back_transcribe()
print(DNAseqB)
```

```python
# Calculate the complementary sequences of DNA sequence A and sequence B
cDNAsA = SeqA.complement()
print(cDNAsA)
cDNAsB = SeqB.complement()
print(cDNAsB)
```

```python
# Calculate similarity of DNA sequence A and sequence B using hamming distance method
def hamming(dnaA,dnaB):
    h = [] #store mismatch
    for x,y in zip(dnaA,dnaB):
      if (x!=y):
          h.append((x,y))
    return len(h)
```

```python
# Map DNA sequence A and sequence B similarity using dot plot technique
def dot_plot(dnaA,dnaB):
    print(" |"+dnaA)
    print("-"*(len(dnaA)+3))
    for b in dnaB:
        print(b+"|", end="")
        for a in dnaA:
            print(" " if a!=b else "X",end="")
        print("")
```
