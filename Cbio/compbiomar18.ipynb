```python
from Bio.Seq import Seq
```

```python
dnaSeq = Seq("ATGATCTCGTAA")
print(dnaSeq)
```

```python
cDNA = dnaSeq.complement()

print(cDNA)
```

```python
mRNA = dnaSeq.transcribe()
print(mRNA)
```

```python
aminoAcidSeq = mRNA.translate()
print(aminoAcidSeq)
```

```python
from Bio.SeqUtils import seq1, seq3
protein3 = seq3(aminoAcidSeq)
print(protein3)
protein1= seq1(protein3)
print(protein1)
```
