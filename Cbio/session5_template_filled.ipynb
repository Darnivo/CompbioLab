# Lab 5 Working with Biological Database

```python
# import Seq
from Bio.Seq import Seq

# import SeqIO
from Bio import SeqIO
```

```python
# clean sequence:
# "BDEFHIJKLMNOPQRSUVWXYZ"
def clean_Sequence(seq):
    for letter in "BDEFHIJKLMNOPQRSUVWXYZ":
        seq = seq.replace(letter,"")
    return seq
```

### Working with Fasta File

```python
# SeqIO.parse()
gb_iterator = SeqIO.parse('sequence.fasta','fasta')
print(gb_iterator)

for record in gb_iterator:
    record.seq = Seq(clean_Sequence(str(record.seq)))
    print(record)
    print(record.id)
    print(record.description)
    print(record.seq)
```

```python
# SeqIO.read()
gb_file = SeqIO.read('sequence.fasta','fasta')
gb_file.seq = Seq(clean_Sequence(str(gb_file.seq)))
print(gb_file)
```
